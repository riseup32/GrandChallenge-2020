```python
import os
import glob

import librosa
import numpy as np
from scipy.io.wavfile import write
```

```python
data_dir = './wav_data/pretrain/KsponSpeech_wav/'
sample_datas = glob.glob(os.path.join(data_dir, '**', '*wav'), recursive=True)
output_dir = './wav_data/normal/'
```

```python
sample_datas
```

```python
y_list = []

for i in range(len(sample_datas)):
    y, sr = librosa.load(sample_datas[0], sr=16000)
    y_list.append(y)
```

```python
y_concat = np.concatenate(y_list)
```
